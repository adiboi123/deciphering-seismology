Importing necessary libraries:

```python
import numpy as np
import matplotlib.pyplot as plt
from random import randint
from bsd import *
```

Creating random data:

```python
t=np.linspace(0,600,1200)
rand_data=np.random.rand(len(t))-0.5
```

```python
fig=plt.figure(figsize=(20,3))
plt.plot(t,rand_data)
plt.xlabel('Time(s)')
plt.ylabel('Amplitude')
plt.title('Generating random signal')
plt.grid()
```

When `plot` is `True`:

```python
blind_source_decon(rand_data,t,plot=True)
```

When `plot` is `False` : *Extracting the optimum wavelet* -

```python
fig=plt.figure(figsize=(10,10))
x=blind_source_decon(rand_data,t,epsilon=1)
plt.plot(-t[:len(x)],x,color='blue')
plt.plot(t[:len(x)],x,color='blue')
plt.ylabel('Amplitude')
plt.xlabel('Time (s)')
plt.title('Optimum Wavelet')
plt.xlim([-10,10]) ## Window used to observe the wavelet
plt.ylim([-20,20])
plt.grid()
```

Comparing optimum wavelets for various $\epsilon$ values for same data:

```python
fig=plt.figure(figsize=(10,10))
for i in range(5):
    x=blind_source_decon(rand_data,t,epsilon=(i+1))
    col='#%06X' % randint(0, 0xFFFFFF)
    plt.plot(-t[:len(x)],x,color=col,label=i+1)
    plt.plot(t[:len(x)],x,color=col)
plt.ylabel('Amplitude')
plt.xlabel('Time (s)')
plt.title('Optimum Wavelets')
plt.xlim([-20,20]) ## Window used to observe the wavelet 
plt.ylim([-15,15])
plt.grid()
plt.legend()
```
